```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

```python
df = pd.read_csv('./brainwave_data/test_RAW_3.csv', usecols = [1])
```

```python
plt.figure(figsize=(12,4))
df.plot()
plt.savefig('brainwave_raw.jpg')
plt.show()
```

```python
L = []
for i in range(128):
    L.append(df.iloc[i])
    
N = 128
T = 1/128
x = np.linspace(0, N*T, N)
yf = fft(L)
xf = np.linspace(0, 1/(2*T), N//2)
plt.plot(xf, abs(yf[0:N//2]), label = 'signal')
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend(loc=1)
```

```python
N = 118
T = 1/118
x = np.linspace(0, N*T, N)
y = df
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2*T), N//2)
plt.plot(xf, abs(yf[0:N//2]), label = 'signal')
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend(loc=1)
```

```python
N = 600
T = 1/600
x = np.linspace(0, N*T, N)
y = np.sin(50*2*np.pi*x)
yf = fft(y)
xf = np.linspace(0, 1/(2*T), N//2)
plt.plot(xf, 2/N*abs(yf[0:N//2]))
plt.show
```
